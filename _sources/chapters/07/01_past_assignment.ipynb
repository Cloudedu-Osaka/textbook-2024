```python
def find_pythagorean_triplets(limit):
    triplets = []
    
    for a in range(1, limit):
        for b in range(a, limit):
            c = (a**2 + b**2)**0.5
            if c.is_integer() and c <= limit:
                triplets.append((a, b, int(c)))
                
    return triplets

limit = 100
triplets = find_pythagorean_triplets(limit)
for triplet in triplets:
    print(triplet)
```

```python
%%ai vioras
$p$, $q$を互いに素な正の整数とすると、
Plimpton 322の最も左のカラムは$(1/2 (p/q + q/p))^2$に対応すると考える事ができます。
このとき、Plimpton 322を生成する$p$と$q$の組み合わせを求めるプログラムを書いてください
```

```python
from math import gcd

def find_plimpton_pairs(limit):
    pairs = []
    
    for p in range(1, limit+1):
        for q in range(1, p):  # ensure p > q
            if gcd(p, q) == 1:  # ensure p and q are coprime
                value = ((p/q + q/p) / 2)**2
                pairs.append(((p, q), value))
    
    return pairs

limit = 50  # adjust this value as needed
pairs = find_plimpton_pairs(limit)
for pair, value in pairs:
    print(f"Pair: {pair}, Value: {value}")
```

### 条件を絞る: $1.4$から$2$の範囲

```python
%%ai vioras
Plimpton 322は、最左のカラムで降順にソートされており、$1.4$から$2$の範囲の値です
```

```python
from math import gcd

def find_plimpton_pairs(limit, min_value, max_value):
    pairs = []
    
    for p in range(1, limit+1):
        for q in range(1, p):  # ensure p > q
            if gcd(p, q) == 1:  # ensure p and q are coprime
                value = ((p/q + q/p) / 2)**2
                if min_value <= value <= max_value:
                    pairs.append(((p, q), value))
    
    # Sort pairs in descending order by value
    pairs.sort(key=lambda x: x[1], reverse=True)
    
    return pairs

limit = 200  # adjust this value as needed
min_value = 1.4
max_value = 2.0
pairs = find_plimpton_pairs(limit, min_value, max_value)
for pair, value in pairs:
    print(f"Pair: {pair}, Value: {value}")
```

### 条件を絞る: 基数の素な約数の合成数

```python
%%ai vioras
さらに、$p$と$q$は$2$,$3$,$5$の合成数と仮定してください
```

```python
from math import gcd

def is_product_of_2_3_5(n):
    for prime in [2, 3, 5]:
        while n % prime == 0:
            n /= prime
    return n == 1

def find_plimpton_pairs(limit, min_value, max_value):
    pairs = []
    
    for p in range(1, limit+1):
        if not is_product_of_2_3_5(p):
            continue
        for q in range(1, p):  # ensure p > q
            if gcd(p, q) == 1 and is_product_of_2_3_5(q):  # ensure p and q are coprime and product of 2, 3, 5
                value = ((p/q + q/p) / 2)**2
                if min_value <= value <= max_value:
                    pairs.append(((p, q), value))
    
    # Sort pairs in descending order by value
    pairs.sort(key=lambda x: x[1], reverse=True)
    
    return pairs

limit = 200  # adjust this value as needed
min_value = 1.4
max_value = 2.0
pairs = find_plimpton_pairs(limit, min_value, max_value)
for pair, value in pairs:
    print(f"Pair: {pair}, Value: {value}")
```

### 数表を復元する

```python
%%ai vioras
Plimpton 322の左から2番目と3番目のカラムは、
$b = p^2 - q^2$、$d = p^2 + q^2$に対応すると考える事ができます。
$b$と$d$も表示して下さい
```

```python
from math import gcd

def is_product_of_2_3_5(n):
    for prime in [2, 3, 5]:
        while n % prime == 0:
            n /= prime
    return n == 1

def find_plimpton_pairs(limit, min_value, max_value):
    pairs = []
    
    for p in range(1, limit+1):
        if not is_product_of_2_3_5(p):
            continue
        for q in range(1, p):  # ensure p > q
            if gcd(p, q) == 1 and is_product_of_2_3_5(q):  # ensure p and q are coprime and product of 2, 3, 5
                value = ((p/q + q/p) / 2)**2
                if min_value <= value <= max_value:
                    b = p**2 - q**2
                    d = p**2 + q**2
                    pairs.append(((p, q, b, d), value))
    
    # Sort pairs in descending order by value
    pairs.sort(key=lambda x: x[1], reverse=True)
    
    return pairs

limit = 200  # adjust this value as needed
min_value = 1.4
max_value = 2.0
pairs = find_plimpton_pairs(limit, min_value, max_value)
for pair, value in pairs:
    print(f"Pair: {pair}, Value: {value}")
```
