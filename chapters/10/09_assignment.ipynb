````python
from sympy import symbols, simplify
from sympy.solvers.inequalities import solve_univariate_inequality

# シンボルを定義
a, b = symbols('a b')

# 不等式を定義
inequality = simplify((a + b)/2 - (a*b)**0.5)

# 不等式を解く
solution = solve_univariate_inequality(inequality >= 0, a)

# 解を出力
print(solution)
````

````python
from sympy import symbols, simplify, solve

# シンボルを定義
a, b = symbols('a b')

# 不等式を定義
inequality = simplify((a + b)/2 - (a*b)**0.5)

# 不等式をaについて解く
solution_a = solve(inequality, a)

# 解を出力
print(solution_a)
````

```{note}
代数的に解けない（式の変形や簡約化では結論に到達しない）問題があります。
その場合は解析的に解く（方程式の根を求めるなど）必要があります。
```

## 参考

````python
import sympy

# a, bを正の実数とする
a, b = sympy.symbols('a b', positive=True)
````

````python
# 相加相乗平均の不等式を表す式
ineq = sympy.sqrt(a * b) <= (a + b) / 2
ineq
````

````python
ineq.lhs
````

````python
ineq.rhs
````

````python
# 不等式を2乗して移項すると、以下の式が得られる
ineq_squared = ineq.rhs**2 - ineq.lhs**2 >= 0
ineq_squared
````

````python
# 左辺を展開すると、以下の式が得られる
lhs_expanded = ineq_squared.lhs.expand()
lhs_expanded
````

````python
sympy.core.assumptions(_)
````

````python
sympy.factor(lhs_expanded)
````
