## Pythonによるプログラミング

### Pythonによる等差数列の表現

```{note}
`$n$`は、数式の変数$n$です。斜体で印字されます。
ChatGPTに数式の変数であることを伝えるためにそのように表記しています。
プログラムの変数は`` `n` ``と表記します。
```

````python
def nth_term(a, d, n):
    return a + (n-1) * d
````

````python
nth_term(1, 1, 100)
````

````python
%%ai vioras
pythonで初項$a$、等差$d$の数列の$n$番目の項を表現することはできますか？
````

````python
def sequence_list(a, d, n):
    return [a + i*d for i in range(n)]
````

````python
sequence_list(1, 1, 10)
````

````python
list(range(1, 11, 1))
````

````python
%%ai vioras
その数列をリストで表現することは？
````

* [List comprehension - Wikipedia](https://en.wikipedia.org/wiki/List_comprehension)

````python
a = 1
d = 1
n = 100
[a + i*d for i in range(n)]
````

````python
(lambda a, d, n: [a + i*d for i in range(n)])(1, 1, 100)
````
